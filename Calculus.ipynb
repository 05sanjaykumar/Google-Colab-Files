```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def df(x):
    return 2*x


x = np.linspace(-10, 10, 400)
y = f(x)
dy = df(x)

plt.figure(figsize=(10, 5))

# Plot function
plt.plot(x, y, label='f(x) = x^2', color='blue')
# Plot derivative
plt.plot(x, dy, label="f'(x) = 2x (slope)", color='red', linestyle='--')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.legend()
plt.title("Function vs Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
#Gradient Descent Step by Step

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x - 3)**2

def df(x):
    return 2 * (x - 3)


# Gradient Descent
x = 0.0                     # Starting point (bad guess!)
lr = 0.1              # Learning rate (step size)
history = [x]

for i in range(20):
  grad = df(x)
  print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}, df(x) = {grad:.4f}")
  x=x -lr*grad #slope upwards +ve and slope downwards -ve
  history.append(x)

# Plotting
x_vals = np.linspace(-1, 5, 100)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label='f(x) = (x-3)^2', color='blue')
plt.scatter(history, [f(i) for i in history], color='red', label='Steps')
plt.plot(history, [f(i) for i in history], color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent Steps Toward Minimum')
plt.legend()
plt.grid(True)
plt.show()
```

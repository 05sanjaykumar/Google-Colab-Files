```python
from scipy.stats import binom

# Parameters
n = 10  # number of trials
p = 0.5  # probability of success
k = 5  # number of successes

# Binomial Probability
prob = binom.pmf(k, n, p)
print(f"Probability of getting exactly {k} heads in {n} tosses:", prob)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# All possible successes from 0 to 10
x = np.arange(0, n+1)
probs = binom.pmf(x, n, p)

# Plot
sns.barplot(x=x, y=probs, color="skyblue")
plt.title("Binomial Distribution: 10 Tosses, p=0.5")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.show()
```

```python
#Binomial probabilty test
arr = []
def putter(n,val):
    if n == 0:
        arr.append(val)
        return
    putter(n-1,val+'H')
    putter(n-1,val+'T')
    return
putter(4,'')
for x in arr:
    print(x)

print('\n probabilty of getting exactly 3 heads')
count = 0
for x in arr:
    temp = 0
    for t in x:
        if t == 'H':
            temp+=1
    if temp == 3:
        count+=1

print(count/len(arr))
```

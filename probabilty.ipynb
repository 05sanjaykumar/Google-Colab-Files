```python
#Basic Definition of probability
import numpy as np

# Probability of an event happening
p_event = 0.7  # 70% chance
p_not_event = 1 - p_event  # Complement rule

print("Probability of event:", p_event)
print("Probability of NOT event:", p_not_event)
```

```python
# Given probabilities
P_A = 0.02  # Probability of having a disease
P_B_given_A = 0.95  # Probability of testing positive if you have the disease
P_B_given_not_A = 0.05  # Probability of testing positive if you don't have the disease
P_not_A = 1 - P_A  # Probability of not having the disease

# Bayes' Theorem calculation
P_A_given_B = (P_B_given_A * P_A) / ((P_B_given_A * P_A) + (P_B_given_not_A * P_not_A))

print("Probability of having the disease given a positive test:", P_A_given_B)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate random data
data = np.random.normal(loc=10, scale=50, size=1000)  # Mean=50, Std Dev=10

# Plot the distribution
sns.histplot(data, bins=30, kde=True)
plt.title("Normal Distribution")
plt.show()

mean = np.mean(data)
variance = np.var(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```
